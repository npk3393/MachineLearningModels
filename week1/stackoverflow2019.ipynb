### Exploratory Data Analysis of StackOverFlow dataset 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("survey_results_public.csv")
```

```python
# view the first 5 records
dataset.head()
```

```python
# Identify the shape of the dataset.
dataset.shape[0],dataset.shape[1]
```

```python
# looking at the metadata as a list
metadatalist = dataset.columns.tolist()
```

```python
for i in len(dataset.columns.tolist()):
    print(dataset.metadatalist[i].value_counts())
```

```python
dataset[metadatalist[3]].value_counts()
```

```python
education = dataset.EdLevel.value_counts()
```

```python
(education/dataset.shape[0]).plot(kind="barh");
plt.title("Formal Education");
```

```python
dataset['EdLevel'].value_counts().plot(kind='barh',figsize=(9,5))
```

```python
dataset['MainBranch'].value_counts()
```

```python
# Horizontal bar plot by extending the value_counts method
dataset['MainBranch'].value_counts().plot(kind='barh', figsize=(9,5))
```

```python
pd.crosstab(dataset.Employment, [dataset.UndergradMajor], rownames=['dataset.Employment'])
```

```python
pd.crosstab(dataset.UndergradMajor, dataset.EdLevel)
```
